```
import tensorflow as tf

#x_train = [1,2,3]
#y_train = [1,2,3]

x_train = tf.placeholder(tf.float32, shape = [None])
y_train = tf.placeholder(tf.float32, shape = [None])

#Variable은 tensorflow가 사용하는 Variable. trainable한 변수  #random_normal() 안에는 shape 지정
w = tf.Variable(tf.random_normal([1]), name = 'weight') 
b = tf.Variable(tf.random_normal([1]), name='bias')

hypothesis = x_train * w + b

#reduce_mean은 텐서가 주어졌을 때 그것을 평균 내 주는 것
cost = tf.reduce_mean(tf.square(hypothesis - y_train))  #squre은 제곱

#cost를 minimize하는 과정
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.01)
train = optimizer.minimize(cost)


sess = tf.Session()
sess.run(tf.global_variables_initializer()) #w,b와 같은 Variable을 사용하기 위해서 써줘야됨

for step in range(2001):
  #sess.run(train)
  #if step % 20 == 0:
  #  print(step, sess.run(cost), sess.run(w), sess.run(b))

  cost_val, w_val, b_val, _ = sess.run([cost, w, b, train], feed_dict={x_train : [1,2,3,4,5], y_train : [2.1,3.1,4.1,5.1,6.1]})
  if step % 20 == 0:
    print(step, cost_val, w_val, b_val)

#학습된 모델 테스트
print(sess.run(hypothesis, feed_dict={x_train : [5]}))
print(sess.run(hypothesis, feed_dict={x_train : [2.5]}))
print(sess.run(hypothesis, feed_dict={x_train : [1.5, 3.5]}))
```
